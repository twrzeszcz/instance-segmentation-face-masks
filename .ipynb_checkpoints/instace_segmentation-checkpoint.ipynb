## Instance Segmentation Project

This is a simple instance segmentation project where we want to segment face masks from images. The dataset used for this can be found here [kaggle](https://www.kaggle.com/andrewmvd/face-mask-detection). We only used images from this dataset and not the annotations cause they were created for detection and not segmentation. Therefore we had to label images ourselves and create bitmap masks. This was done using a *Labelme* software which allow to create polygon points around the objects. Then masks and bounding boxes are automatically generated. We have labelled a subset of 300 images for training and 60 images for testing which can be found on the github repo for this project here [Github](https://github.com/twrzeszcz/instance-segmentation-face-masks). You can find a tutorial for labelling here [medium](https://olafenwaayoola.medium.com/image-annotation-with-labelme-81687ac2d077).

For training we gonna use the pretrained Mask RCNN model from the *pixellib* package. After training we will run the inference on images and also segment a video. All the steps for training a custom model and inference can be found here [Github1](https://github.com/ayoolaolafenwa/PixelLib/blob/master/Tutorials/custom_train.md), [Github2](https://github.com/ayoolaolafenwa/PixelLib/blob/master/Tutorials/custom_inference.md).

```python
import pixellib
from pixellib.custom_train import instance_custom_training
from pixellib.instance import custom_segmentation
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

Training was done on the Google Colab to speed it up. As the first step we have to download a pretrained Mask RCNN model (link in the *Github1* tutorial). I have used both a *resnet50* and *resnet101* backbone. With resnet50 the smallest validation loss was roughly 0.8 at the epoch 9. For resnet101 we could achieve the loss of roughly 0.55 at the epoch 22. Training for more epochs did not improve the validation loss but only the training loss. So labelling more images would definitely help to improve the model. Before training, images are also augmented with a left-right, upside-down flips and a gaussian blur as a default.

```python
train_maskrcnn = instance_custom_training()
train_maskrcnn.modelConfig(network_backbone='resnet50', num_classes=1, batch_size=6)
train_maskrcnn.load_pretrained_model('mask_rcnn_coco.h5')
train_maskrcnn.load_dataset('face_masks')
train_maskrcnn.train_model(num_epochs=100, augmentation=True, path_trained_models='mask_rcnn_models_resent50')
```

```python
train_maskrcnn = instance_custom_training()
train_maskrcnn.modelConfig(network_backbone='resnet101', num_classes=1, batch_size=4)
train_maskrcnn.load_pretrained_model('mask_rcnn_coco.h5')
train_maskrcnn.load_dataset('face_masks')
train_maskrcnn.train_model(num_epochs=100, augmentation=True, path_trained_models='mask_rcnn_models_resent101')
```

Now we can load the best model from all saved checkpoints and evaluate it using a mAP(Mean Average Precision) metric.

```python
train_maskrcnn = instance_custom_training()
train_maskrcnn.modelConfig(network_backbone='resnet50', num_classes=1)
train_maskrcnn.load_dataset('face_masks')
train_maskrcnn.evaluate_model('mask_rcnn_models_resnet50/mask_rcnn_model_resnet50.h5')
```

```python
train_maskrcnn = instance_custom_training()
train_maskrcnn.modelConfig(network_backbone='resnet101', num_classes=1)
train_maskrcnn.load_dataset('face_masks')
train_maskrcnn.evaluate_model('mask_rcnn_models_resnet101/mask_rcnn_model_resnet101.h5')
```

The resnet101 model did better as it was expected. Lets now run the inference to see if this difference is visible. When we specify classes for inference, the first one always has to be the background denoted as *'BG'*.

### Resnet50 backbone

```python
segmentation_model_resnet50 = custom_segmentation()
segmentation_model_resnet50.inferConfig(num_classes=1, class_names=['BG', 'mask'], 
                                       network_backbone='resnet50', detection_threshold=0.8)
segmentation_model_resnet50.load_model('mask_rcnn_models_resnet50/mask_rcnn_model_resnet50.h5')
```

```python
for img_path in ['maksssksksss96.png', 'maksssksksss81.png', 'maksssksksss450.png', 'maksssksksss451.png']:
    segmask, output = segmentation_model_resnet50.segmentImage('train_unused/' + img_path, show_bboxes=True)
    plt.figure(figsize=(10,10))
    plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
    plt.show()
```

### Resnet101 backbone

```python
segmentation_model_resnet101 = custom_segmentation()
segmentation_model_resnet101.inferConfig(num_classes=1, class_names=['BG', 'mask'], 
                                       network_backbone='resnet101', detection_threshold=0.8)
segmentation_model_resnet101.load_model('mask_rcnn_models_resnet101/mask_rcnn_model_resnet101.h5')
```

```python
for img_path in ['maksssksksss96.png', 'maksssksksss81.png', 'maksssksksss450.png', 'maksssksksss451.png']:
    segmask, output = segmentation_model_resnet101.segmentImage('train_unused/' + img_path, show_bboxes=True)
    plt.figure(figsize=(10,10))
    plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
    plt.show()
```

It looks like both models are very accurate and give similar results. Before we move to the real time and video segmentation lets also see how we can extract segmented objects.

```python
segmask, output = segmentation_model_resnet101.segmentImage('train_unused/maksssksksss81.png', 
                                                            show_bboxes=True, extract_segmented_objects=True)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(output, cv2.COLOR_BGR2RGB))
for ext_object in segmask['extracted_objects']:
    plt.figure(figsize=(5,5))
    plt.imshow(cv2.cvtColor(ext_object, cv2.COLOR_BGR2RGB))
```

Real time segmentation can be done very easily using a *segmentFrame* function which modifies every frame in place. However to run the inference smoothly you need a very powerful GPU and on the CPU you will get crazy low fps. That is why I have tried the segmentation on the prerecorded video.

```python
capture = cv2.VideoCapture(0)
while True:
    ret, frame = capture.read()
    segmentation_model_resnet50.segmentFrame(frame)
    cv2.imshow('frame', frame)
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
```

I have used 3 different videos from here [mixkit](https://mixkit.co/free-stock-video/mask/). Segmented versions are uploaded on the github repo for this project here [Github](https://github.com/twrzeszcz/instance-segmentation-face-masks).

```python
segmask, output = segmentation_model_resnet101.process_video("video_sample_3.mp4", show_bboxes=True,
                            output_video_name="video_sample_seg_3.mp4", frames_per_second=30)
```

One can also record a video using a webcam and then apply a segmentation model on it. Below is the code for recording a webcam stream.

```python
cap = cv2.VideoCapture(0)

fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 30.0, (640,480))

while True:
    ret, frame = cap.read()
    if ret==True:
        out.write(frame)

        cv2.imshow('frame', frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
        
cap.release()
out.release()
cv2.destroyAllWindows()
```
